**Assignment-1**

**Puntuation Removal,Stopwords Removal,stemmimg and Lemmatization**

```python
import nltk
nltk.download('punkt_tab')
```

```python
text="""A smartphone is a sophisticated handheld device that has revolutionized
the way people communicate,work, and entertain themselves. Combining the features
of a mobile phone, camera, computer, and GPS into a single compact device,
smartphones have become an essential tool in modern life."""
```

**Punctuation Removal**

```python
import re
text_clean=re.sub(r'[^\w\s]','',text)
print(text_clean)
```

**Tokenization**

```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text_clean.upper())
print(tokens)
```

**Stopwords** **Removal**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
filtered_words=[word for word in tokens if word.upper()not in stop_words]
print(filtered_words)
```

**Stemming**

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_words=[stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

**Lemmatization**

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered_words]
print(lemmatized_words)
```

```python
coverted_text=" ".join(lemmatized_words)
print(coverted_text)
```

**Word2Vec**

```python
Large_text="""Social media refers to digital platforms and online applications that enable people to create, share,
and exchange content while connecting with others around the world. Platforms such as Facebook, Instagram, Twitter,
LinkedIn, and TikTok have completely transformed the way individuals and organizations communicate, share information,
and build communities. Over the past two decades, social media has grown from a simple networking tool into one of
the most powerful forces shaping modern culture, politics, and business. Companies leverage social media to market
their products, engage with customers, and build brand loyalty, while individuals use it to express opinions,
share experiences, and stay informed about current events. It has played a vital role in driving social movements,
raising awareness about important issues, and giving a voice to marginalized communities across the globe. News and
information now spread within seconds, reaching millions of people instantly. However, social media also carries
significant drawbacks, including the rapid spread of misinformation, cyberbullying, data privacy concerns, and its
damaging effects on mental health, especially among teenagers and young adults. Studies have linked excessive social
media use to anxiety, depression, and low self-esteem. Despite these challenges, social media continues to evolve and
remains an inseparable part of modern human life."""
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(large_text.lower())
print(sentences)
```

```python
import re
clean_sentences = [re.sub(r'[^\w\s]', '', sentence) for sentence in sentences]
print(clean_sentences)
```

```python
tokens_1 = [word_tokenize(sentence) for sentence in clean_sentences]
print(tokens_1)
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec
model=Word2Vec(sentences=tokens_1,
               vector_size=100,
               window=3,
               min_count=1,
               epochs=50,)
```

```python
print("\nSimilarity between 'facebook' and 'tiktok':")
print(model.wv.similarity('facebook', 'tiktok'))
```

```python
print("\nSimilarity between 'social' and 'media':")
print(model.wv.similarity('social', 'media'))
```

```python
print("\nSimilarity between 'business' and 'companies':")
print(model.wv.similarity('business', 'companies'))
```
